```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
accident_data = pd.read_csv("accident.csv")
accident_data
```

```python
accident_data.columns
```

```python
MAN_COLL = accident_data["MAN_COLL"].value_counts()
FATALS = accident_data["FATALS"].value_counts()
```

```python
accident_data = MAN_COLL.add(FATALS, fill_value=0)
accident_data
```

```python
accident_data.plot(kind="bar", figsize=(20,10))
plt.title("Manner of Collision vs Fatalities")
plt.ylabel("Number of Fatalies")
plt.xlabel("Manner of Collision")
plt.xticks(np.arange(11), ('Not a Collision with a Motor Vehicle In-Transport ', 'Front-to-Rear', 'Front-to-Front', 'Angle', 'Sideswipe-Same Direction', 'Sideswipe-Oppostie Direction', 'Rear-to-Side', 'Rear-to-Rear', 'Other', 'Not Reported', 'Unknown'))
```
