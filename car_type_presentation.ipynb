```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
all_df = pd.read_csv("Resources/person.csv")
all_df.head()
all_df.columns
```

```python
def_df = pd.read_csv("Resources/body_type.csv")
def_df.head()
#def_df.count()
# all_df.columns
```

```python
# Get body_type data - all
body_type_data = all_df["BODY_TYP"].value_counts()
body_type_data.sort_values(ascending=False)
first_20_types = body_type_data.nlargest(20)

first_20_types.values

body_type_df = pd.DataFrame({'BODY_TYPE' : first_20_types.index, 
    'accident_total' : first_20_types.values
})

body_type_df

final_data = pd.merge(body_type_df, def_df, on = 'BODY_TYPE', how='inner')
final_data.count()
```

```python
# Configure plot and ticks
first_20_types.plot(kind="bar", facecolor="red",rot = 85)
```

```python
final_data
final_data['Definition']
final_data['accident_total']
final_data.count()
```

```python

plt.figure(figsize=(30,10))
plt.title("Body Types Accidents")
plt.xlabel("Body Type")
plt.ylabel("Number of Accedents")
plt.bar(final_data['Definition'], final_data['accident_total'] , facecolor="red", alpha=0.75,align="center")
plt.xticks(rotation=90)
plt.show()


```
