```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
accident_data = pd.read_csv("accident.csv")
accident_data
```

```python
accident_data.columns
```

```python
State_Deaths = accident_data.loc[:, ["STATE","ARR_MIN","FATALS"]]
State_Deaths


State_Deaths = State_Deaths.set_index("STATE")

State_Deaths.head() 
```

```python

CO_Deaths = State_Deaths.loc[6,["ARR_MIN","FATALS"]]
CO_Deaths
```

```python
ARR_MIN = CO_Deaths["ARR_MIN"].value_counts()
FATALS = CO_Deaths["FATALS"].value_counts()
```

```python
CO_Deaths = ARR_MIN.add(FATALS, fill_value=0)
CO_Deaths
```

```python
CO_Deaths.plot(kind="bar", figsize=(20,10))
plt.title("Arrival Time of EMS vs Fatalities")
plt.ylabel("FATALS")
plt.xlabel("Arrival Time of EMS in Min")
plt.show()
```
